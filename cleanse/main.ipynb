```python
import pandas as pd
```

```python
file_path = '../RawData/surgery_indicator.csv'
surgery_data = pd.read_csv(file_path)

surgery_data.head()
```

```python
surgery_data_cleaned = surgery_data.dropna(axis=0, how='all').dropna(axis=1, how='all')

header_row_index = None
for idx, row in surgery_data_cleaned.iterrows():
    if idx > 0 and not surgery_data_cleaned.iloc[idx - 1].isna().all():
        header_row_index = idx - 1
        break

header_row_index, surgery_data_cleaned.head(100)
```

```python
surgery_data_cleaned = pd.read_csv(file_path, header=1)

surgery_data_cleaned = surgery_data_cleaned.dropna(thresh=surgery_data_cleaned.shape[1] - 2)

surgery_data_cleaned.head(100)
```

```python
surgery_data_cleaned = surgery_data_cleaned.iloc[1:].dropna(axis=1, how='all')

column_names = {
    'Unnamed: 0': 'Indicator',
    'Unnamed: 1': 'Target',
    'Unnamed: 2': 'Year_2561',
    'Unnamed: 3': 'Year_2562',
    'Unnamed: 4': 'Year_2563',
    'Unnamed: 5': 'Year_2564',
    'Unnamed: 6': 'Year_2565',
    'Unnamed: 7': 'Year_2566',
}
surgery_data_cleaned = surgery_data_cleaned.rename(columns=column_names)

surgery_data_cleaned.head()
```

```python
missing_values = surgery_data_cleaned.isna().sum()
missing_values
```

```python
surgery_data_cleaned.head()
```

```python
from googletrans import Translator

translator = Translator()

try:
    surgery_data_cleaned['Indicator'] = surgery_data_cleaned['Indicator'].apply(
        lambda x: translator.translate(x, src='th', dest='en').text
    )
except AttributeError as e:
    print(f"An error occurred: {e}")

surgery_data_cleaned.head(100)
```

```python
unique_target_values = surgery_data_cleaned['Target'].dropna().unique()
unique_target_values
```

```python
surgery_data_cleaned['Target'].fillna(-1, inplace=True)

surgery_data_cleaned.head()
```

```python
surgery_data.to_csv("clean.csv")
```
